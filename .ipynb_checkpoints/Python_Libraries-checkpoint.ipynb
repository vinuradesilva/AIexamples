```python
#Multiple curves in single plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for multiple curves
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x / 5

# Plot each curve
plt.plot(x, y1, label='Sine wave', color='blue', linestyle='-')
plt.plot(x, y2, label='Cosine wave', color='red', linestyle='--')
plt.plot(x, y3, label='Linear function', color='green', linestyle=':')

# Add labels and title for clarity
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Curves on One Diagram')

# Add a legend to identify each curve
plt.legend()

# Display the plot
plt.show()
```

```python
#Piece-wise functions
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define the domain
x = np.linspace(-5, 5, 500) #from -5 to +5, with 500 points

#defining the piecewise function
f = np.piecewise(
    x,
    [x<0, (x>=0) & (x<2), x>=2],
    [lambda x: -x, lambda x: x**2, lambda x: 2*x + 1]
)

#Plotting
plt.plot(x, f, label="F(x)")
plt.title("Piecewise Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
